```python
import numpy as np
import matplotlib.pyplot as plt

# Data
labels = ["Accuracy", "F1-Score", "Kappa", "AUC-ROC"]
models = {
    "Stacking (No Optuna)": [0.8902, 0.86621, 0.81641, 0.94739],
    "Stacking (Optuna - LogReg)": [0.8646, 0.8365, 0.7736, 0.9355],
    "Stacking (Optuna - ExtraTrees)": [0.8642, 0.8362, 0.7729, 0.9352]
}

line_styles = ['solid', 'dashed', 'dotted']
markers = ['o', 's', '^']
colors = ['tab:blue', 'tab:orange', 'tab:green']

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for i, (model, values) in enumerate(models.items()):
    values += values[:1]  
    ax.plot(angles, values, label=model, linestyle=line_styles[i], marker=markers[i], color=colors[i])
    ax.fill(angles, values, alpha=0.1, color=colors[i])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_yticklabels([])
ax.set_title("Optimized Stacking Classifier Performance", size=14)
ax.legend(loc="upper right", bbox_to_anchor=(1.4, 1))

plt.tight_layout()
plt.show()
```
